```python
import tkinter as tk
from tkinter import *
from tkinter import ttk
from tkinter import messagebox
from tkinter import filedialog 
from time import strftime
import csv

### Tạo các biến trung gian để lưu dữ liệu
dat_todo = []
dat_doing = []
dat_done = []
dat_merge = []

def savedtb():
    for items in todo_tree.get_children():
        dat_todo.append(todo_tree.item(items)['values'])

### Chức năng tạo treeview
def create_tree(root):
    columns = ('Group', 'Task')
    tree = ttk.Treeview(root, columns=columns, show='headings')
    tree.heading('Group', text='Group')
    tree.heading('Task', text='Task')
    tree.column("Group", anchor=tk.CENTER, width=90)
    tree.column("Task",  width=280)  
    return tree

### Chức năng chọn task trên treeview và hiện thị thông tin của task đang chọn trong entry box 
def OnClick1(e):
    # Reset entry box
    group.delete(0, 'end')
    task.delete(0, 'end')
    task_description.delete(0, 'end')
    # Lấy dữ liệu từ task đang lựa chọn (double click chuột trái) trên treeview
    selected_todo = todo_tree.selection()
    values_todo = todo_tree.item(selected_todo, 'values')
    # Hiển thị dữ liệu của task đang chọn vào trường nhập dữ liệu
    group.insert(0, values_todo[0])
    task.insert(0, values_todo[1])
    task_description.insert(0, values_todo[2])
def OnClick2(e):    
    # Lấy dữ liệu từ task đang lựa chọn (click chuột trái) trên treeview
    selected_doing = doing_tree.selection()
    values_doing = doing_tree.item(selected_doing, 'values')
    # Hiển thị dữ liệu của task đang chọn vào trường nhập dữ liệu
    group.insert(0, values_doing[0])
    task.insert(0, values_doing[1])
    task_description.insert(0, values_doing[2])
def OnClick3(e):
    # Lấy dữ liệu từ task đang lựa chọn (click chuột trái) trên treeview
    selected_done = done_tree.selection()
    values_done = done_tree.item(selected_done, 'values')
    # Hiển thị dữ liệu của task đang chọn vào trường nhập dữ liệu
    group.insert(0, values_done[0])
    task.insert(0, values_done[1])
    task_description.insert(0, values_done[2])

### Chức năng cập nhật file csv lưu trữ dữ liệu trên máy tính
def save_todo():
    savedtb()
    with open('todo_list.csv', mode="w", newline='') as myfile:
        writer = csv.writer(myfile, delimiter=',')
        for i in todo_tree.get_children():
            row = todo_tree.item(i)['values']
            row.append(day_string)
            writer.writerow(row)
def save_doing():
    with open('doing_list.csv', mode="w", newline='') as myfile:
        writer = csv.writer(myfile, delimiter=',')
        for i in doing_tree.get_children():
            row = doing_tree.item(i)['values']
            row.append(day_string)
            writer.writerow(row)
def save_done():
    with open('done_list.csv', mode="w", newline='') as myfile:
        writer = csv.writer(myfile, delimiter=',')
        for i in done_tree.get_children():
            row = done_tree.item(i)['values']
            row.append(day_string)
            writer.writerow(row)
            
### Chức năng thêm task vào trong treeview
def add():
    if len(group.get())==0 & len(task.get())==0:
        pass
    else:
    # Hiển thị thông tin task trong treeview với thông tin vừa nhập ở entry box
        todo_tree.insert('', 'end', values=(group.get(), task.get(), task_description.get()))
    # Reset entry box
        group.delete(0, 'end')
        task.delete(0, 'end')
        task_description.delete(0, 'end')
    # Update file csv dữ liệu trên máy tính
        save_todo()

### Chức năng xóa task đang lựa chọn (click chuột trái) trên treeview
def delete():
    respone = messagebox.askyesno(title='Confirmation',
                       message='Bạn có muốn xóa công việc đang lựa chọn?')
    selected_item_todo = todo_tree.selection()
    if respone == 1:
        for items in selected_item_todo:
            todo_tree.delete(items)  
            
    selected_item_doing = doing_tree.selection()
    if respone == 1:
        for items in selected_item_doing:
            doing_tree.delete(items)  
            
    selected_item_done = done_tree.selection()
    if respone == 1:
        for items in selected_item_done:
            done_tree.delete(items)  
    # Reset entry box
    group.delete(0, 'end')
    task.delete(0, 'end')
    task_description.delete(0, 'end')
    # Update file csv dữ liệu trên máy tính
    save_todo()
    save_doing()
    save_done()

### Chức năng xóa toàn bộ task đang hiển thị trên treeview
def clear():
    respone = messagebox.askyesno(title='Confirmation',
                       message='Bạn có muốn xóa toàn bộ dữ liệu?')
    if respone == 1:
        for items in todo_tree.get_children():
            todo_tree.delete(items)
        for items in doing_tree.get_children():
            doing_tree.delete(items)
        for items in done_tree.get_children():
            done_tree.delete(items)
    # Reset entry box
    group.delete(0, 'end')
    task.delete(0, 'end')
    task_description.delete(0, 'end')
    # Update file csv dữ liệu trên máy tính
    save_todo()
    save_doing()
    save_done()

### Chức năng chỉnh sửa task đang hiển thị trên treeview
def edit():
    selected_item_todo = todo_tree.selection()
    # Thay thế thông tin vừa nhập vào task đang lựa chọn trên treeview
    todo_tree.item(selected_item_todo, text="",
              values=(group.get(), task.get(), task_description.get()))
    
    selected_item_doing = doing_tree.selection()
    # Thay thế thông tin vừa nhập vào task đang lựa chọn trên treeview
    doing_tree.item(selected_item_doing, text="",
              values=(group.get(), task.get(), task_description.get()))
    
    selected_item_done = done_tree.selection()
    # Thay thế thông tin vừa nhập vào task đang lựa chọn trên treeview
    done_tree.item(selected_item_done, text="",
              values=(group.get(), task.get(), task_description.get()))
    
    # Reset entry box
    group.delete(0, 'end')
    task.delete(0, 'end')
    task_description.delete(0, 'end')
    # Update file csv dữ liệu trên máy tính
    save_todo()
    save_doing()
    save_done()

### Chức năng thoát chương trình
def exit():
    respone = messagebox.askyesno(title='Confirmation', 
                       message='Bạn có muốn thoát khỏi chương trình không?')
    if respone == 1:
        root.destroy()  

### Chức năng chuyển đổi task qua lại giữa các danh mục
def move12():
    if len(group.get())!=0:
        selected_item = todo_tree.selection()
        values = todo_tree.item(selected_item, 'values')
        doing_tree.insert('', 'end', values=values)
        todo_tree.delete(selected_item)
        # Reset entry box
        group.delete(0, 'end')
        task.delete(0, 'end')
        task_description.delete(0, 'end')
        # Update file csv dữ liệu trên máy tính
        save_todo()
        save_doing()
    else:
        messagebox.showwarning("Waring", "Phải lựa chọn 1 task trong Todo-list")
        
def move21():
    if len(group.get())!=0:
        selected_item = doing_tree.selection()
        values = doing_tree.item(selected_item, 'values')
        todo_tree.insert('', 'end', values=values)
        doing_tree.delete(selected_item)
        # Reset entry box
        group.delete(0, 'end')
        task.delete(0, 'end')
        task_description.delete(0, 'end')
        # Update file csv dữ liệu trên máy tính
        save_todo()
        save_doing()
    else:
        messagebox.showwarning("Waring", "Phải lựa chọn 1 task trong Doing-list")
        
def move23():
    if len(group.get())!=0:
        selected_item = doing_tree.selection()
        values = doing_tree.item(selected_item, 'values')
        done_tree.insert('', 'end', values=values)
        doing_tree.delete(selected_item)
        # Reset entry box
        group.delete(0, 'end')
        task.delete(0, 'end')
        task_description.delete(0, 'end')
        # Update file csv dữ liệu trên máy tính
        save_doing()
        save_done()
    else:
        messagebox.showwarning("Waring", "Phải lựa chọn 1 task trong Doing-list")
        
def move32():
    if len(group.get())!=0:
        selected_item = done_tree.selection()
        values = done_tree.item(selected_item, 'values')
        doing_tree.insert('', 'end', values=values)
        done_tree.delete(selected_item)
        # Reset entry box
        group.delete(0, 'end')
        task.delete(0, 'end')
        task_description.delete(0, 'end')
        # Update file csv dữ liệu trên máy tính
        save_doing()
        save_done()
    else:
        messagebox.showwarning("Waring", "Phải lựa chọn 1 task trong Done-list")

### Hiển thị thời gian hiện tại
time_string = strftime('%H:%M:%S %p \n %A \n %d %b, %Y') # time format 
day_string = strftime('%d/%b/%Y') # time format 

### Chức năng sắp xếp danh mục task theo Group
def sort():
    row_todo = [(todo_tree.set(item, 'Group'), item) for item in todo_tree.get_children('')]
    row_todo.sort()
    # rearrange items in sorted positions
    for index, (values, item) in enumerate(row_todo):
        todo_tree.move(item, '', index)
    
    row_doing = [(doing_tree.set(item, 'Group'), item) for item in doing_tree.get_children('')]
    row_doing.sort()
    # rearrange items in sorted positions
    for index, (values, item) in enumerate(row_doing):
        doing_tree.move(item, '', index)
        
    row_done = [(done_tree.set(item, 'Group'), item) for item in done_tree.get_children('')]
    row_done.sort()
    # rearrange items in sorted positions
    for index, (values, item) in enumerate(row_done):
        done_tree.move(item, '', index)

my_data = []
def import_csv():           
    with open('todo_list.csv') as myfile:
        reader = csv.DictReader(myfile, delimiter=',')
        for row in reader:
            my_data.append(row)
```

```python
### Man hinh chinh
root = tk.Tk()
root.title('Daily Task Management')
root.geometry("1228x430")
root.resizable(0,0)

### Tạo treeview danh sách task (ToDo, Doing, Done) và nút chuyển task giữa các treeview
task_frame = ttk.LabelFrame(root)
task_frame.grid(row=0, column=0, padx=2, pady=2, sticky='nw')

todo_lab = ttk.Label(task_frame, text='ToDo', font=('Calibri', 12, 'bold'))
todo_lab.grid(row=0, column=0)
todo_tree = create_tree(task_frame)
todo_tree.grid(row=1, column=0, padx = 2, pady = 2)

frm_move1 = tk.Frame(task_frame)
frm_move1.grid(row=1, column=1)
right_move1 = ttk.Button(frm_move1, text=">>", width=5, command=move12)
right_move1.grid(row=0, column=0, padx=2)
left_move1 = ttk.Button(frm_move1, text="<<", width=5, command=move21)
left_move1.grid(row=1, column=0, padx=2)

doing_lab = ttk.Label(task_frame, text='Doing', font=('Calibri', 12, 'bold'))
doing_lab.grid(row=0, column=2)
doing_tree = create_tree(task_frame)
doing_tree.grid(row=1, column=2, padx=2, pady=2)

frm_move2 = tk.Frame(task_frame)
frm_move2.grid(row=1, column=3)
right_move2= ttk.Button(frm_move2, text=">>", width=5, command=move23)
right_move2.grid(row=0, column=0, padx=2)
left_move2 = ttk.Button(frm_move2, text="<<", width=5, command=move32)
left_move2.grid(row=1, column=0, padx=2)

done_lab = ttk.Label(task_frame, text='Done', font=('Calibri', 12, 'bold'))
done_lab.grid(row=0, column=4)
done_tree = create_tree(task_frame)
done_tree.grid(row=1, column=4, padx=2, pady=2)

### Tạo khung nhập/sửa task
bottom_frame = tk.Frame(root)
bottom_frame.grid(row=1, column=0, padx=2, pady=2, sticky='nw')

input_frame = ttk.LabelFrame(bottom_frame, text='Add/Edit the Task')
input_frame.grid(row=0, column=0, padx=2, pady=2, sticky='nw')

group_lab = ttk.Label(input_frame, text='Group:')
group_lab.grid(row=0, column=0, padx=5, pady=5, sticky='nw')
group = ttk.Entry(input_frame, width=30)
group.grid(row=0, column=1, padx=5, pady=5, sticky='nw')

task_lab = ttk.Label(input_frame, text='Task:')
task_lab.grid(row=1, column=0, padx=5, pady=5, sticky='nw')
task = ttk.Entry(input_frame, width=30)
task.grid(row=1, column=1, padx=5, pady=5, sticky='nw')

task_description_lab = ttk.Label(input_frame, text='Description:')
task_description_lab.grid(row=0, column=2, padx=2, pady=2, sticky='nw')
task_description = ttk.Entry(input_frame, width=111)
task_description.grid(row=0, column=3, padx=2, pady=2)

add_btn = ttk.Button(input_frame, text='Add', command=add)
add_btn.grid(row=0, column=4, padx=5, pady=2)
edit_btn = ttk.Button(input_frame, text='Edit', command=edit)
edit_btn.grid(row=1, column=4, padx=5, pady=2)

### Tạo các nút chức năng
command_frame = ttk.LabelFrame(bottom_frame, text='Command')
command_frame.grid(row=1, column=0, padx=2, pady=2, sticky='nw')

del_btn = ttk.Button(command_frame, text='Remove task', command=delete)
del_btn.grid(row=0, column=0, ipadx=5, padx=5, pady=5)

clear_btn = ttk.Button(command_frame, text='Clear all', command=clear)
clear_btn.grid(row=0, column=1, padx=5, pady=5)

sort_btn = ttk.Button(command_frame, text='Sort by Group', command=sort)
sort_btn.grid(row=0, column=2, ipadx=5, padx=5, pady=5)

import_btn = ttk.Button(command_frame, text='Import')
import_btn.grid(row=0, column=3, padx=5, pady=5)

export_btn = ttk.Button(command_frame, text='Export')
export_btn.grid(row=0, column=4, padx=5, pady=5)

exit_btn = ttk.Button(command_frame, text='Exit', command=exit)
exit_btn.grid(row=0, column=5, padx=5, pady=5)

### Hiển thị thời gian hiện tại
time_frame = ttk.LabelFrame(bottom_frame)
time_frame.grid(row=0, column=1, padx=10, columnspan=2)

time = tk.Label(time_frame, font=('Calibri', 14, 'bold'))
time.config(text=time_string)
time.grid(row=0, column=0, sticky='n')

### Hiển thị thông tin task trong entry box khi lựa chọn trong treeview
todo_tree.bind('<ButtonRelease-1>', OnClick1)
doing_tree.bind('<ButtonRelease-1>', OnClick2)
done_tree.bind('<ButtonRelease-1>', OnClick3)

### Chay chuong trinh
root.mainloop()
```
